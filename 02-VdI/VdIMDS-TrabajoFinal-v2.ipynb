## Grafico 1 .- Cumplimiento de OF

```python
a = p2.groupby('Orden Fab.')['Cantidad OF','Cant. Piso'].mean().reset_index()
a
```

```python
def porcentaje (cantof, cantpiso):
    porcentaje = cantpiso/cantof*100
    return porcentaje
a['cumplimiento']=a[['Cantidad OF','Cant. Piso']].apply(lambda x : porcentaje(*x), axis=1)
a
```

```python
porcentaje_ordenado=a.sort_values(by='cumplimiento') #ver qué tan critica es la OF (tiempo paa cumplirla)
porcentaje_ordenado
```

```python
top10=porcentaje_ordenado.iloc[0:10]
top10=pd.DataFrame(top10)
top10
```

```python
top10.set_index('Orden Fab.', inplace=True)
```

```python
df_sorted = top10['cumplimiento'].sort_values(ascending=False)
df_sorted =pd.DataFrame(df_sorted)
df_sorted['pendiente'] = 100-df_sorted
```

```python
#Función que permite eliminar los bordes del gráfico
def hide_border(ax):
    ax.spines['bottom'].set_visible(False) #Linea inferior
    ax.spines['left'].set_visible(False) #Linea izquierda
    ax.spines['right'].set_visible(False) #Linea derecha
    ax.spines['top'].set_visible(False) #Linea superior
```

```python
fontsize=12
colors = ["#36a7e9", '#d3d3d3']

ax = df_sorted.plot.barh(stacked=True, figsize=(10, 5), legend=False, width=0.9, alpha=0.9, xlim=(0, 100),
                        color=colors)

hide_border(ax)   #Ocultar bordes
ax.tick_params(axis=u'both', which=u'both', length=0)
ax.set_ylabel('')
ax.set_title("Top 10 de #OF con menor % de cumplimiento", pad=24, fontweight=700)
ax.title.set_position([.5, 1])
percents = df_sorted.columns
ax.text(0, df_sorted.shape[0]-0.3, #ubicacion horizontal, posicion vertica,
        percents[0], color=colors[0], fontsize=fontsize)  #nombre de la variable, color del texto, tamaño texto 
ax.text(85, df_sorted.shape[0]-0.3, percents[1], color=colors[1], fontsize=fontsize)

for p in ax.patches:
    if p.xy[0] == 0:
        xpos = 2
    else:
        xpos = 93
    ax.annotate(str(int(round(p.get_width()))) + "%", (xpos, p.xy[1] + 0.3), color='black', fontsize=fontsize)

plt.show()
```
