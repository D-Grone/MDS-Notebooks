## 2.- Análisis Exploratorio

### 2.1.- Cálculo de Frecuencia

Lo primero es identificar las variables No Categoricas

```python
cols = df.columns
num_cols_no_categoric = df._get_numeric_data().columns
num_cols_no_categoric
```

Se identifican las variables Categóricas

```python
col_categoric = list(set(cols) - set(num_cols_no_categoric))
col_categoric
```

Se cuantifican las observaciones de cada clase (frecuencia)

```python
frecuencia_data =pd.value_counts(df[col_categoric].values.flatten())
sns.catplot(x="classes", kind="count", palette="ch:.25", data=df)
```

### 2.2.- Análisis de Correlaciones

```python
fig, (ax) = plt.subplots(1, 1, figsize=(10,6))
corr = df[num_cols_no_categoric].corr()

hm = sns.heatmap(corr, ax=ax, cmap="coolwarm", annot=True, fmt='.2f', linewidths=0)

fig.subplots_adjust(top=0.93)
fig.suptitle(' Mapa de Calor de Correlación de Vinos', fontsize=14, fontweight='bold')
```
