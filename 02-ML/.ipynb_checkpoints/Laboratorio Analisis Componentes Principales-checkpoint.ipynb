```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import fetch_california_housing

california = fetch_california_housing()
```

```python
import pandas as pd
df = pd.DataFrame(california.data,columns=california.feature_names)
```

```python
df.head()
```

```python
X_train, x_test, y_train, y_test = train_test_split(df,california.target,test_size=0.3,random_state=123)
```

```python
X_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
pca = PCA(.95)
```

```python
scaler = StandardScaler()
pcafit = pca.fit(scaler.fit_transform(X_train))
```

```python
pcafit.n_components_
```

```python
import matplotlib.pyplot as plt
import numpy as np

PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
loadings = pca.transform(x_test)
pca.transform(x_test)
```

```python
import matplotlib.pyplot as plt 
import numpy as np
 
# Get the loadings of x and y axes
xs = loadings[0]
ys = loadings[1]
 
# Plot the loadings on a scatterplot
for i, varnames in enumerate(california.feature_names[0:6]):
    plt.scatter(xs[i], ys[i], s=200)
    plt.arrow(
        0, 0, # coordinates of arrow base
        xs[i], # length of the arrow along x
        ys[i], # length of the arrow along y
        color='r', 
        head_width=0.01
        )
    plt.text(xs[i], ys[i], varnames)
 
# Define the axes
#xticks = np.linspace(-0.8, 0.8, num=5)
#yticks = np.linspace(-0.8, 0.8, num=5)
#plt.xticks(xticks)
#plt.yticks(yticks)
plt.xlabel('PC1')
plt.ylabel('PC2')
 
# Show plot
plt.title('2D Loading plot with vectors')
plt.show()
```

# Kmeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(df).score(df) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
# Predicting the clusters
labels = kmeans.predict(df)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df['mean radius'], df["mean texture"], df["mean perimeter"], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df['mean radius'], df["mean area"], df["mean smoothness"], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```

# Clustering Jerarquico

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
```

```python
labels = hierarchical_cluster.fit_predict(df) 
```

```python
plt.scatter(df['mean radius'], df['mean smoothness'], c=labels)
plt.show()
```

```python

from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linkage_data = linkage(df, method='ward', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(df) 
plt.scatter(df['mean radius'], df['mean smoothness'], c=labels)
plt.show()
```

```python
colores=['red','green','blue','cyan']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df['mean smoothness'], df["mean compactness"], df["mean symmetry"], c=asignar,s=60)
```
